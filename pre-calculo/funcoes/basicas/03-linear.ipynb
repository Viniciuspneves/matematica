# O que é uma Função Linear?

A função linear é uma das mais importantes da matemática e modela situações onde há uma relação constante entre duas variáveis. Sua forma geral é:

$$
f(x) = ax + b
$$

- $a$ é o coeficiente angular (indica a inclinação da reta).
- $b$ é o coeficiente linear (indica onde a reta corta o eixo $y$).

Ela descreve retas não verticais, e aparece em muitos contextos do mundo real, como cálculos de custo, velocidade constante, variações de temperatura, entre outros.

---

## Como ela é escrita?

A forma geral da função linear é:

$$
f(x) = ax + b
$$

- Se $a = 0$, a função vira uma constante (reta horizontal).
- Se $b = 0$, a reta passa pela origem.
- A função sempre gera uma reta, com inclinação definida por $a$.

### Exemplo:

- Se $f(x) = 2x + 3$, então:
  - $f(0) = 3$
  - $f(1) = 5$
  - $f(-2) = -1$

---

## Características

- Gráfico: Reta inclinada que pode subir ou descer.
- Crescimento:
  - Crescente se $a > 0$
  - Decrescente se $a < 0$
- Raiz: O ponto onde a função cruza o eixo $x$, obtido por:

$$
f(x) = 0 \Rightarrow x = -\frac{b}{a}
$$

- Domínio: $\mathbb{R}$
- Imagem: $\mathbb{R}$

---

## E o gráfico?

O gráfico da função linear é uma reta. Sua inclinação depende do valor de $a$, e sua altura inicial depende de $b$.

Por exemplo, se $f(x) = 2x + 1$, a reta começa em $y = 1$ e sobe 2 unidades para cada 1 unidade que avança em $x$.

---
## Aplicações no dia a dia

Funções lineares aparecem em diversos contextos que envolvem crescimento ou decrescimento constante.

| Situação                                      | Explicação                                      | Função linear        |
|----------------------------------------------|------------------------------------------------|----------------------|
| Cálculo de custo fixo + variável             | R\$10 fixo + R\$2 por item                        | $f(x) = 2x + 10$       |
| Salário com valor por hora + adicional fixo  | R\$20 por hora + R\$100 fixo                      | $f(x) = 20x + 100$     |
| Temperatura aumentando 2º por hora           | A cada hora a temperatura sobe 2 graus          | $f(x) = 2x + T₀$       |
| Distância percorrida em velocidade constante | Um carro andando a 60 km/h                      | $f(x) = 60x$           |
| Valor de um serviço com taxa inicial         | R\$30 fixo + R\$5 por km de deslocamento          | $f(x) = 5x + 30$       |


## Exemplo aplicado: Custo de produção linear

### Enunciado

Uma gráfica cobra R$5,00 por impressão colorida. Não há taxa adicional fixa, ou seja, o custo depende apenas da quantidade de impressões.

1. Modelando a função que representa o custo total em função da quantidade de impressões realizadas.
2. Determinando o custo para 20, 40, 60, 80 e 100 impressões.
3. O gráfico da função no intervalo de $x = 0$ até $x = 100$.

---

### Solução

#### 1. Modelagem da função

Como o custo é proporcional à quantidade de impressões, temos uma função linear do tipo:

$$
f(x) = 5x
$$

Onde:
- $x$ = quantidade de impressões
- $f(x)$ = custo total em reais

---

#### 2. Aplicando valores

- Para 20 impressões: $f(20) = 5 \cdot 10 = R\$50,00$
- Para 40 impressões: $f(40) = 5 \cdot 25 = R\$200,00$
- Para 60 impressões: $f(60) = 5 \cdot 50 = R\$300,00$
- Para 80 impressões: $f(80) = 5 \cdot 75 = R\$400,00$
- Para 100 impressões: $f(100) = 5 \cdot 100 = R\$500,00$

---

#### 3. Gráfico

O gráfico é uma reta que passa pela origem $(0, 0)$ e cresce com coeficiente angular $a = 5$, indicando que a cada impressão o custo aumenta R$5,00.



```python
import matplotlib.pyplot as plt
import numpy as np

# Intervalo de x (quantidade de impressões)
x = np.linspace(0, 100, 100)  # de 0 a 100, com 100 pontos
y = 5 * x  # função linear: custo = 5 * quantidade

# Criando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x) = 5x', color='steelblue', linewidth=2)

# Destaques nos pontos do enunciado
quantidades = [20, 40, 60, 80, 100]
custos = [5 * q for q in quantidades]
plt.scatter(quantidades, custos, color='orangered', zorder=5, linewidth=0.5)
for xi, yi in zip(quantidades, custos):
    plt.annotate(f"f({xi}), R${yi})", (xi, yi), textcoords="offset points", xytext=(0,5), ha='right')

# Estilização
plt.title("Custo de Impressões Coloridas (R$5,00 cada)")
plt.xlabel("Quantidade de Impressões (x)")
plt.ylabel("Custo Total (R$)")
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```
