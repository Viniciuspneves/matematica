# O que é uma Função Identidade?

A função identidade é uma das funções mais fundamentais da matemática. Ela é definida por:

$$
f(x) = x
$$

Ou seja, o valor de saída é sempre igual ao valor de entrada. Se você coloca 2, ela retorna 2. Se coloca -5, ela retorna -5. Ela "espelha" o valor que você insere — por isso o nome "identidade".

Seu gráfico é uma reta diagonal que passa pela origem $(0, 0)$ e sobe com inclinação de 45°.

Apesar de parecer simples, a função identidade é a base para entender transformações lineares, composições de funções e o comportamento de diversos sistemas reais em que a variação é direta e proporcional.

---

## Como ela é escrita?

A forma geral é:

$$
f(x) = x
$$

- O valor de entrada x é exatamente igual ao valor de saída f(x).
- Essa função é uma espécie de "espelho": o que entra é o que sai.

### Exemplo:

- Se $x = 3$, então $f(3) = 3$  
- Se $x = -7$, então $f(-7) = -7$

---

## Características

- O gráfico é uma reta passando pela origem.
- Cresce de forma constante — quanto maior o $x$, maior o $f(x)$.
- A taxa de crescimento é 1, ou seja, cada unidade que aumenta em $x$, também aumenta em $f(x)$.
- Domínio: $\mathbb{R}$
- Imagem: $\mathbb{R}$
- Raiz: $x = 0$

---

## E o gráfico?

O gráfico da função identidade é uma reta que passa por $(0, 0)$ e cresce uniformemente.

É a linha que divide o plano em dois: tudo o que está acima dela representa $f(x) > x$ e abaixo, $f(x) < x$ (quando comparada com outras funções).

---

## Aplicações no dia a dia

A função identidade aparece em situações onde o valor de saída é proporcional ao valor de entrada, sem alteração. Aqui vão alguns exemplos em níveis de dificuldade:

| Situação                                | Explicação                                                 | Função identidade |
|----------------------------------------|-------------------------------------------------------------|-------------------|
| Contagem simples de objetos            | 1 item → 1 ponto; 5 itens → 5 pontos                        | $f(x) = x$        |
| Cálculo de salário por hora (R$1/h)    | Cada hora trabalhada vale exatamente R$1,00                 | $f(x) = x$        |
| Conversão 1:1 (ex: polegada para polegada) | A medida em polegadas permanece igual                     | $f(x) = x$        |
| Intensidade de som igual à entrada     | Em um sistema linear sem amplificação                       | $f(x) = x$        |
| Modelo de reflexão no espelho          | A imagem refletida segue exatamente a forma real            | $f(x) = x$        |


## Exemplo aplicado: Controle de Velocidade em Esteira de Produção

### Funcionamento Correto (sem falha)

Uma esteira de produção foi ajustada para funcionar de forma que a velocidade real da esteira seja igual à velocidade configurada no painel, em metros por minuto. Assim, temos a seguinte função identidade:

$$
v(s) = s
$$

Onde:

- $v(s)$ é a velocidade real da esteira;
- $s$ é a velocidade configurada pelo operador no painel de controle.

Isso significa que qualquer valor definido no painel será exatamente a velocidade que a esteira vai executar. Essa correspondência direta representa um sistema ideal, eficiente e previsível.

### Exemplos práticos:
- Configurou 40 $m/min$ → esteira opera a 40 $m/min$.  
- Configurou 60 $m/min$ → esteira opera a 60 $m/min$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Valores para o gráfico
s = np.linspace(0, 100, 400)

# Função identidade correta
v = s

# Criando os gráficos
plt.figure(figsize=(10, 6))

# Linha de referência
plt.axline((0, 0), slope=1, color='gray', linestyle=':', linewidth=1)

# Linha funcionamento correto
plt.plot(s, v, label='Funcionamento Correto: $v(s) = s$', color='steelblue', linewidth=2)

# Configurações do gráfico
plt.title('Controle de Velocidade da Esteira de Produção')
plt.xlabel('Velocidade Configurada (m/min)')
plt.ylabel('Velocidade Real (m/min)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Exemplo aplicado: Controle de Velocidade em Esteira de Produção

### Funcionamento com (Falha)

Suponha agora que haja uma falha no sistema de controle da esteira. Por exemplo, um defeito nos sensores de velocidade faz com que a esteira opere com 5 metros por minuto a mais do que o valor configurado no painel.

A função que representa esse comportamento é:

$$
v_{\text{falha}}(s) = s + 5
$$

Onde:

- $v_{\text{falha}}(s)$ é a velocidade real da esteira em falha;
- $s$ é a velocidade configurada no painel.

Essa pequena diferença entre o valor configurado e o valor real pode causar problemas graves na produção, como falhas no empacotamento, produtos caindo fora da posição ou máquinas descompassadas.

### Exemplos práticos:
- Configurou 40 $m/min$ → esteira opera a 45 $m/min$.  
- Configurou 60 $m/min$ → esteira opera a 65 $m/min$.

Essa função mostra a importância de sistemas de controle confiáveis, onde a identidade $f(x) = x$ é preservada.


```python
import numpy as np
import matplotlib.pyplot as plt

# Valores para o gráfico
s = np.linspace(0, 100, 400)

# Função identidade correta
v_correto = s

# Função com falha (erro de +5 m/min)
v_falha = s + 5

# Criando os gráficos
plt.figure(figsize=(10, 6))

# Linha de referência
plt.axline((0, 0), slope=1, color='gray', linestyle=':', linewidth=1)

# Linha funcionamento correto/falha
plt.plot(s, v_correto, label='Funcionamento Correto: $v(s) = s$', color='steelblue', linewidth=2)
plt.plot(s, v_falha, label='Com Falha: $v(s) = s + 5$', color='orangered', linestyle='--', linewidth=2)


# Configurações do gráfico
plt.title('Controle de Velocidade da Esteira de Produção')
plt.xlabel('Velocidade Configurada (m/min)')
plt.ylabel('Velocidade Real (m/min)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
