# O que é uma Função Quadrática?

A função quadrática é uma função polinomial de segundo grau, muito comum em problemas de otimização, trajetórias e física. Sua forma geral é:

$$
f(x) = ax^2 + bx + c
$$

Onde:
- $a \neq 0$
- $a, b, c \in \mathbb{R}$

---

## Como ela é escrita?

A forma padrão da função quadrática é:

$$
f(x) = ax^2 + bx + c
$$

- $a$: **coeficiente quadrático**, responsável pela concavidade.  
- $b$: **coeficiente linear**, influencia a simetria e a posição da parábola.  
- $c$: **coeficiente constante**, indica o ponto onde a parábola cruza o eixo $y$ (ou seja, $f(0) = c$).

### Exemplo:

Se $f(x) = 2x^2 - 4x + 1$, então:

- $f(0) = 1$  
- $f(1) = -1$  
- $f(2) = 1$

---

## Características

- **Gráfico**: Uma **parábola**.  
  - Se $a > 0$: concavidade voltada para cima (mínimo).  
  - Se $a < 0$: concavidade voltada para baixo (máximo).  

- **Vértice**: Ponto mais alto ou mais baixo da parábola, calculado por:

$$
x_v = \frac{-b}{2a}, \quad y_v = f(x_v)
$$

- **Eixo de simetria**: Linha vertical que passa pelo vértice: $x = x_v$.

- **Raízes (ou zeros)**: Valores de $x$ que tornam $f(x) = 0$, calculados pela fórmula de Bhaskara:

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

- **Domínio**: $\mathbb{R}$  
- **Imagem**:  
  - Se $a > 0$: $[y_v, +\infty)$  
  - Se $a < 0$: $(-\infty, y_v]$

---

## E o gráfico?

O gráfico da função quadrática é uma **parábola**.

- O vértice é o ponto de mínimo (se $a > 0$) ou máximo (se $a < 0$).  
- A parábola é **simétrica** em relação ao eixo vertical que passa pelo vértice.  
- Pode ter:
  - Duas raízes reais (corta o eixo $x$ em dois pontos);  
  - Uma raiz real (toca o eixo $x$ num único ponto);  
  - Nenhuma raiz real (não cruza o eixo $x$).

---

## Aplicações no dia a dia

As funções quadráticas aparecem em diversas situações onde há **variação não linear**, especialmente em movimentos, áreas e otimizações.

| Situação                                      | Explicação                                                                 | Função quadrática         |
|----------------------------------------------|---------------------------------------------------------------------------|----------------------------|
| Lançamento de projéteis                      | A altura varia com o tempo de forma parabólica                           | $f(t) = -5t^2 + 20t$       |
| Lucro em função do preço                     | Aumentar o preço demais reduz as vendas, criando um pico de lucro        | $f(x) = -2x^2 + 40x$       |
| Área máxima com perímetro fixo               | Otimizar formas (ex: retângulo)                                           | $A(x) = -x^2 + 20x$        |
| Curvas de aceleração ou frenagem             | A posição varia quadraticamente com o tempo                               | $f(t) = at^2 + vt + s$     |
| Economia de combustível                      | Rendimento de carro pode variar de forma parabólica com a velocidade     | $f(v) = -0{,}1v^2 + 5v$    |


## Exercício — Trajetória de um foguete 

Durante um teste de lançamento, a altura (em metros) de um pequeno foguete em função do tempo (em segundos) após o disparo é dada por:

$$
h(t) = -4.9t^2 + 39.2t
$$

Essa equação representa a trajetória vertical do foguete, considerando apenas a força da gravidade (sem resistência do ar).

---

### Perguntas:

1. Qual é a altura máxima que o foguete atinge? Em que instante isso ocorre?  
2. Após quantos segundos o foguete volta ao solo?
3. Represente graficamente a altura em função do tempo.
4. Qual a altura do foguete após 1 segundo? E após 3 segundos?

---

### Observações:

Essa é uma função quadrática do tipo:

$$
h(t) = at^2 + bt + c
$$

onde:  
- \( a = -4.9 \)  
- \( b = 39.2 \)  
- \( c = 0 \)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

# Coeficientes da equação quadrática
a = -4.9
b = 39.2
c = 0

# Valores de x no intervalo desejado
x = np.linspace(-0.5, 8.5, 100)  # Gera 100 pontos de -0.5 a 8,5

# Calcula os valores de y
y = a * x**2 + b * x + c

# Calcula o vértice
x_vertice = -b / (2 * a)
y_vertice = a * x_vertice**2 + b * x_vertice + c

# Calcula as raízes (x1 e x2)
delta = b**2 - 4 * a * c
x1 = (-b + np.sqrt(delta)) / (2 * a)
x2 = (-b - np.sqrt(delta)) / (2 * a)

# Calcula a altura do foguete após 1 e 3 segundos
x_vertice_1_seg = 1 # valores ja definidos
x_vertice_3_seg = 3

# Subistituimos na equação
y_vertice_1_seg = a * x_vertice_1_seg**2 + b * x_vertice_1_seg + c
y_vertice_3_seg = a * x_vertice_3_seg**2 + b * x_vertice_3_seg + c

# Cria o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x, y, label=rf"$f(x) = ax^2 + bx + c$", color="orangered", zorder=3)

# Eixos
plt.axhline(y=0, color='black', linewidth=1.2)  # Eixo X
plt.axvline(x=0, color='black', linewidth=1.2)  # Eixo Y

# Atualiza os limites para usar nas proporções
y_min, y_max = plt.ylim()
x_min, x_max = plt.xlim()

# Calcula proporções para destacar o vértice corretamente
ymax_prop = (y_vertice - y_min) / (y_max - y_min)
xmax_prop = (x_vertice - x_min) / (x_max - x_min)

# Linhas tracejadas até o vértice
plt.axvline(x=x_vertice, color='tomato', linestyle="--", linewidth=1,
            ymin=0, ymax=ymax_prop)
plt.axhline(y=y_vertice, color='tomato', linestyle="--", linewidth=1,
            xmin=0, xmax=xmax_prop)

# Marca o vértice
plt.scatter(x_vertice, y_vertice, color="dodgerblue", zorder=5, label="Vértice")

# Marca os segundos
plt.scatter(x_vertice_1_seg, y_vertice_1_seg, color="limegreen", zorder=5, label=rf"$V_1seg$")
plt.scatter(x_vertice_3_seg, y_vertice_3_seg, color="deeppink", zorder=5, label=rf"$V_3seg$")
plt.text(x_vertice_1_seg + 0.1, y_vertice_1_seg - 1, rf"$V_1({y_vertice_1_seg:.0f})m$", fontsize=8)
plt.text(x_vertice_3_seg + 0.1, y_vertice_3_seg - 1, rf"$V_1({y_vertice_1_seg:.0f})m$", fontsize=8)

# Marca as raízes
plt.scatter([x1, x2], [0, 0], color="purple", zorder=5, label="$Raízes (x_1, x_2)$")
plt.text(x_vertice - 2.5, y_vertice - 2, rf"$_V({y_vertice:.0f})m$", fontsize=8)

# Respondendo as perguntas
print(f"A altura máxima atingida é {y_vertice} m ocorrendo no segundo {x_vertice:.0f} ")
print(f"O foguete volta ao solo após {x2:.0f} segundos")
print(f"O foguete atinge a altura de {y_vertice_1_seg:.2f} m após 1 segundo")
print(f"E a altura de {y_vertice_3_seg:.2f} m após 3 segundos")

# Legendas
plt.title("Gráfico tragetória do foguete", fontsize=14)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

## (Enem 2016)
Um túnel deve ser lacrado com uma tampa de concreto. A seção transversal do túnel e a tampa de concreto têm contornos de um arco de parábola e mesmas dimensões. Para determinar o custo da obra, um engenheiro deve calcular a área sob o arco parabólico em questão. Usando o eixo horizontal no nível do chão e o eixo de simetria da parábola como eixo vertical, obteve a seguinte equação para a parábola:   
$y = 9 – x^2$ , sendo $x$ e $y$ medidos em metros.  

Sabe-se que a área sob uma parábola como esta é igual a $\frac{2}{3}$ da área do retângulo cujas dimensões são, respectivamente, iguais à base e à altura da entrada do túnel.   

Qual é a área da parte frontal da tampa de concreto, em metro quadrado?   

- a) 18
- b) 20
- c) 36
- d) 45
- e) 54

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

# Coeficientes da equação quadrática
a = -1
b = 0
c = 9

# Valores de x no intervalo desejado
x = np.linspace(-3.2, 3.2, 100)
y = a * x**2 + b * x + c

# Calcula o vértice
x_vertice = -b / (2 * a)
y_vertice = a * x_vertice**2 + b * x_vertice + c

# Calcula as raízes (x1 e x2)
delta = b**2 - 4 * a * c
x1 = (-b + np.sqrt(delta)) / (2 * a)
x2 = (-b - np.sqrt(delta)) / (2 * a)

# Calcula a área
frac = 2/3  #fornecido no enunciado
base = abs(x1 -x2)
altura = c

area = frac * base * altura

# Cria o gráfico
plt.figure(figsize=(10, 8))
plt.plot(x, y, label=rf"$f(x) = ax^2 + bx + c$", color="orangered", zorder=3)

# Eixos
plt.axhline(y=0, color='black', linewidth=1.2)  # Eixo X
plt.axvline(x=0, color='black', linewidth=1.2)  # Eixo Y

# Atualiza os limites para usar nas proporções
y_min, y_max = plt.ylim()
x_min, x_max = plt.xlim()

# Calcula proporções para destacar o vértice corretamente
ymax_prop = (y_vertice - y_min) / (y_max - y_min)
xmax_prop = (x_vertice - x_min) / (x_max - x_min)

# Linhas tracejadas até o vértice
plt.axvline(x=x_vertice, color='tomato', linestyle="--", linewidth=1,
            ymin=0, ymax=ymax_prop)
plt.axhline(y=y_vertice, color='tomato', linestyle="--", linewidth=1,
            xmin=0, xmax=xmax_prop)

# Marca o vértice
plt.scatter(x_vertice, y_vertice, color="slateblue", zorder=5, label=rf"$Vértice$")
plt.text(x_vertice - 2, y_vertice - 0.3, rf"$_V({y_vertice:.2f})$", fontsize=8)

# Marca as raízes
plt.scatter([x1, x2], [0, 0], color="purple", zorder=5, label= rf"$Raízes (x_1, x_2)$")
plt.text(x1, -0.7, rf"$x_1 ({x1:.2f})$", fontsize=8)
plt.text(x2 - 0.4, -0.7, rf"$x_2 ({x2:.2f})$", fontsize=8)

# Preenche a área sob a curva
plt.fill_between(x, y, where=(y > 0), color="orange", alpha=0.2, label=rf"$Área$")
plt.text(x_vertice + 0.5, y_vertice - 5, rf"$Área({area:.0f})m^2$", fontsize=8)

# Respondendo a pergunta
print(f"A área da parte frontal da tampa de concreto é a resposta C {area:.2f} m²")

# Legendas
plt.title("Gráfico área da parábola", fontsize=14)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
