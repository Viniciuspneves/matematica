# O que é uma Função Cúbica?

A **função cúbica** é uma função polinomial de grau três. Sua forma geral é:

$$
f(x) = ax^3 + bx^2 + cx + d
$$

Onde:
- $a \neq 0$
- $a, b, c, d \in \mathbb{R}$

---

## Como ela é escrita?

A forma padrão da função cúbica é:

$$
f(x) = ax^3 + bx^2 + cx + d
$$

- $a$: **coeficiente cúbico**, determina o crescimento ou decrescimento dominante da função.  
- $b$: **coeficiente quadrático**, influencia a curvatura e possíveis pontos de inflexão.  
- $c$: **coeficiente linear**, afeta a inclinação local da curva.  
- $d$: **coeficiente constante**, é o valor de $f(0)$ (intercepta o eixo $y$).

### Exemplo:

Se $f(x) = x^3 - 3x^2 - x + 3$, então:

- $f(0) = 3$  
- $f(1) = 0$  
- $f(2) = -1$  
- $f(3) = 0$

---

## Características

- **Grau**: 3 (função do terceiro grau).  
- **Gráfico**: Uma **curva contínua**, geralmente com **um ponto de inflexão**. Pode ter **1 ou 3 raízes reais**.

### Comportamento:

- Se $a > 0$: a função **desce à esquerda** e **sobe à direita**.  
- Se $a < 0$: a função **sobe à esquerda** e **desce à direita**.


###  Raízes:

A função cúbica pode ter:
- Uma única raiz real (e duas complexas),
- Ou até três raízes reais distintas.

Para encontrá-las, podemos usar:
- Fatoração (se possível),
- Gráfico,
- Substituição ou métodos numéricos.

---

## E o gráfico?

O gráfico da função cúbica é uma **curva suave e contínua**. Pode apresentar:

- **2 extremos locais** (um máximo e um mínimo),  
- **1 ponto de inflexão** (onde a curvatura muda).

A depender dos coeficientes, o gráfico pode ter formas bem variadas, mas sempre apresentará crescimento ou decrescimento global de acordo com o sinal de $a$.

---

## Aplicações no dia a dia

As funções cúbicas aparecem em contextos onde há **variações mais complexas e assimétricas**, como:

| Situação                                     | Explicação                                                                 | Função cúbica                    |
|---------------------------------------------|-----------------------------------------------------------------------------|----------------------------------|
| Crescimento populacional não linear         | Crescimento acelerado seguido de desaceleração                             | $f(t) = t^3 - 6t^2 + 9t$         |
| Lucro com tendência sazonal                 | Modela variações de vendas com pontos críticos                             | $f(x) = -x^3 + 6x^2 - 9x + 4$    |
| Curvas de aprendizado                       | Representa avanço rápido seguido de estabilização                          | $f(t) = at^3 + bt^2 + ct + d$    |
| Otimização com mais de um ponto crítico     | Permite encontrar máximos e mínimos locais                                 | $f(x) = x^3 - 3x$                |
| Engenharia e controle de sistemas           | Comum em análises de comportamento dinâmico                                | $f(x) = ax^3 + bx^2 + cx + d$    |


---

## Como encontrar as raízes de uma função cúbica?

As **raízes** (ou **zeros**) de uma função cúbica $f(x) = ax^3 + bx^2 + cx + d$ são os valores de $x$ que satisfazem:

$$
f(x) = 0
$$

Ou seja:

$$
ax^3 + bx^2 + cx + d = 0
$$

Ao contrário das funções quadráticas (que têm a fórmula de Bhaskara), **não há uma fórmula simples e direta** para resolver qualquer função cúbica. Porém, existem **métodos práticos** que você pode aplicar, dependendo do caso:


---

### Teorema do Racional (ou Teorema de Girard-Ruffini)

Tente **testar divisores do termo constante $d$** para ver se algum valor é raiz:

Exemplo:

Se $f(x) = x^3 - 6x^2 + 11x - 6$, os divisores de $d = -6$ são:

$$
\pm1, \pm2, \pm3, \pm6
$$

Testando $x = 1$:

$$
f(1) = 1 - 6 + 11 - 6 = 0 \Rightarrow x = 1 \text{ é raiz!}
$$

Agora podemos fazer a **divisão polinomial** ou usar **Briot-Ruffini** para simplificar a equação para uma função quadrática, e resolver com Bhaskara.

---

### Método de Briot-Ruffini (Divisão Sintética)

Quando você encontra uma raiz $r$, pode dividir o polinômio original por $(x - r)$ usando esse método.

Depois de reduzir a função cúbica a uma função quadrática, basta aplicar a fórmula de Bhaskara para encontrar as outras duas raízes.

---


## Resumo:
  
Se encontrar uma raiz, use **Ruffini** ou **divisão polinomial** para reduzir a equação.  
Resolva a equação quadrática resultante com Bhaskara.

## Exercício - Análise de Lucro com Função Cúbica (Ano Fiscal)

### Contexto

Uma empresa modelou seu lucro trimestral usando a função:

$$
f(x) = -x^3 + 6x^2 - 9x + 4
$$

onde:

- $f(x)$ representa o Lucro **(centenas de milhares de reais)**
**  
- $x$ representa o **tempo em trimestres** após o início da operação

---

### Objetivo

- Entender como o lucro varia ao longo do **ano fiscal** ($0 \leq x \leq 4$)
- Identificar pontos de máximo e mínimo
- Visualizar o gráfico e interpretar tendências

---

### Resolução Matemática

**Pontos críticos:**

- $x = 1$: mínimo local  
- $x = 3$: máximo local

**Análise do comportamento:**

| Intervalo     | Sinal de $f'(x)$ | Comportamento     |
|---------------|------------------|--------------------|
| $0 < x < 1$   | negativo         | decrescente      |
| $1 < x < 3$   | positivo         | crescente        |
| $3 < x < 4$   | negativo         | decrescente      |

---

### Conclusão

- O lucro **cai até o 1º trimestre**, sobe até o **3º trimestre**, e depois volta a cair.
- O **lucro máximo** ocorre no **3º trimestre**.
- Isso pode indicar uma **tendência sazonal** no desempenho financeiro da empresa.


```python
import numpy as np
import matplotlib.pyplot as plt

# Eixo x: de 0 a 4 trimestres
x = np.linspace(0, 4, 400)

# Função cúbica (lucro)
y = -x**3 + 6*x**2 - 9*x + 4

# Pontos críticos
x_crit = np.array([1, 3])
y_crit = -x_crit**3 + 6*x_crit**2 - 9*x_crit + 4

# Gráfico
plt.figure(figsize=(10,6))

# Eixos
plt.axhline(y=0, color='black', linewidth=1.2)  # Eixo X
plt.axvline(x=0, color='black', linewidth=1.2)  # Eixo Y

plt.plot(x, y, label=r'$f(x) = -x^3 + 6x^2 - 9x + 4$', color='indigo')
plt.scatter(x_crit, y_crit, color='navy', zorder=5, label='Pontos críticos', linewidths=0.5)

# Anotações
plt.text(1.05, y_crit[0]-0.4, f'x = 1\nLucro = (R${y_crit[0]:.2f})', ha='left', color='black', fontsize=8)
plt.text(3.05, y_crit[1] + 0.1, f'x = 3\nLucro = (R${y_crit[1]*100:.3f})', ha='left', color='black', fontsize=8)

# Preenche a área sob a curva
plt.fill_between(x, y, where=((y > 0) & (x > 1) & (x < 3)), color="mediumseagreen", alpha=0.2, label="Lucro")
plt.fill_between(x, y, where=((y > 0) & ((x < 1) | (x > 3))), color="mediumvioletred", alpha=0.2, label="Queda")

# Legendas
plt.title('Lucro da empresa ao longo dos trimestres (1 ano)')
plt.xlabel('Tempo (trimestres)')
plt.ylabel('Lucro (centenas de milhares de reais)')
plt.xlim(0, 4)
plt.ylim(min(y)-2, max(y)+2)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
