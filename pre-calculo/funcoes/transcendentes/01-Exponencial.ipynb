# O que é uma Função Exponencial?

A **função exponencial** é caracterizada pela **variável estar no expoente**. Ela modela crescimentos ou decaimentos que ocorrem a uma **taxa constante sobre um período**.

Sua forma geral é:

$$
f(x) = C \cdot a^x
$$

Onde:
- $C$ é uma constante real ($\neq 0$)
- $a$ é a base, uma constante real **positiva** e **diferente de 1** ($a > 0$ e $a \neq 1$)
- $x$ é a variável independente (real)

---

## Como ela é escrita?

A forma padrão da função exponencial é:

$$
f(x) = C \cdot a^x
$$

- $C$: **coeficiente inicial** ou **valor inicial**. É o valor de $f(0)$, pois $f(0) = C \cdot a^0 = C \cdot 1 = C$. Representa o ponto onde o gráfico intercepta o eixo $y$.
- $a$: **base da potência**. É o fator pelo qual a função multiplica seu valor a cada incremento unitário em $x$. Determina se a função **cresce** ou **decresce** e a **rapidez** dessa variação.

### Exemplo:

Se $f(x) = 3 \cdot 2^x$, então:

- $C = 3$ (o gráfico corta o eixo $y$ em 3)
- $a = 2$ (a base)

Calculando alguns pontos:
- $f(0) = 3 \cdot 2^0 = 3 \cdot 1 = 3$
- $f(1) = 3 \cdot 2^1 = 3 \cdot 2 = 6$
- $f(2) = 3 \cdot 2^2 = 3 \cdot 4 = 12$
- $f(-1) = 3 \cdot 2^{-1} = 3 \cdot \frac{1}{2} = 1.5$

---

## Características

- **Domínio**: Geralmente, todos os números reais ($\mathbb{R}$).
- **Imagem**: Depende do sinal de $C$. Se $C > 0$, a imagem é $(0, +\infty)$. Se $C < 0$, a imagem é $(-\infty, 0)$. **Nunca atinge o zero**.
- **Gráfico**: Uma **curva contínua**, sempre **crescente** ou sempre **decrescente**. Possui uma **assíntota horizontal**.

### Comportamento:

- Se $a > 1$: a função é **sempre crescente**. O gráfico sobe da esquerda para a direita. (Ex: $2^x$, $3^x$)
- Se $0 < a < 1$: a função é **sempre decrescente**. O gráfico desce da esquerda para a direita. (Ex: $(1/2)^x$, $(0.8)^x$)

### Raízes:

A função exponencial básica $f(x) = C \cdot a^x$ **não possui raízes reais**. Isso significa que o gráfico **nunca cruza o eixo $x$**. Ele se aproxima dele (a assíntota horizontal), mas nunca o toca.

---

## E o gráfico?

O gráfico da função exponencial é uma **curva suave e contínua**. Ele sempre passa pelo ponto $(0, C)$.

- **Não possui máximos ou mínimos locais**.
- Possui uma **assíntota horizontal** (a linha que o gráfico se aproxima infinitamente). Para $f(x) = C \cdot a^x$, a assíntota horizontal é o eixo $x$, ou seja, a reta $y = 0$.

A forma do gráfico é determinada pela base $a$ e pelo sinal de $C$:

- Se $a > 1$ e $C > 0$: Cresce rapidamente à direita, aproximando-se de $y=0$ à esquerda.
- Se $0 < a < 1$ e $C > 0$: Decresce rapidamente à esquerda, aproximando-se de $y=0$ à direita.
- Se $a > 1$ e $C < 0$: Decresce à direita (para valores negativos), aproximando-se de $y=0$ (por baixo) à esquerda.
- Se $0 < a < 1$ e $C < 0$: Cresce à esquerda (para valores negativos), aproximando-se de $y=0$ (por baixo) à direita.

---

## Aplicações no dia a dia

As funções exponenciais são essenciais para modelar fenômenos onde a **taxa de mudança é proporcional ao valor atual**, levando a crescimentos ou decaimentos muito rápidos (exponenciais):

| Situação                                  | Explicação                                                                 | Função Exponencial (Forma comum) |
|-------------------------------------------|-----------------------------------------------------------------------------|----------------------------------|
| **Crescimento Populacional** | Modelagem inicial em ambientes ideais sem restrições                        | $P(t) = P_0 \cdot a^t$ ou $P_0 \cdot e^{kt}$ |
| **Juros Compostos** | O dinheiro cresce sobre o próprio juro acumulado                           | $M(t) = P \cdot (1+i)^t$         |
| **Decaimento Radioativo** | Quantidade de material radioativo diminui pela metade em intervalos fixos (meia-vida) | $N(t) = N_0 \cdot (1/2)^{t/T}$ ou $N_0 \cdot e^{-\lambda t}$ |
| **Propagação de Doenças/Vírus (início)** | Número de infectados pode crescer muito rapidamente                         | $I(t) = I_0 \cdot a^t$           |
| **Resfriamento de Corpos** | (Lei de Resfriamento de Newton) A diferença de temperatura diminui exponencialmente | $\Delta T(t) = \Delta T_0 \cdot e^{-kt}$ |
| **Descarga de Capacitores** | A carga elétrica diminui exponencialmente ao longo do tempo                | $Q(t) = Q_0 \cdot e^{-t/RC}$     |

### Exemplo

Em um tipo de epidemia, cientistas afirmam que daqui a $"t"$ dias a quantidade de bactérias em (milhões) no ar segue a seguinte função 
$$
f(t) = 100 \cdot e^{0,02t}
$$

Qual será o número de bactérias em 1 e em 10 dias?

```python
import numpy as np

def f(t):
    return 100 * np.exp(0.02 * t)
```

```python
print(f"O numero de bacterias em 1 dia é {f(1)}")
print(f"O numero de bacterias em 10 dias é {f(10)}")
```

### Gráfico Enunciado

- **Observação:**

O gráfico aparenta ter o shape de uma função linear por conta dos valores terem um intervalo curto, conforme o enunciado do exercício, mas a seguir dá para ver a forma da função plotada no gráfico.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme()

# Função exponencial
def f(t):
    return 100 * np.exp(0.02 * t)

# Gráfico da função
plt.figure(figsize=(12, 6))

# Limites do Gráfico

# Definindo o intervalo de t
t = np.linspace(0, 10, 11)
plt.xlim(t[0]-1, t[-1]+1)
plt.ylim(f(t[0]-1), f(t[-1]+1))  # Aumenta o limite superior para dar espaço ao gráfico
ax = sns.lineplot(x=t, y=f(t), label=r'$f(t) = 100 \cdot e^{0.02t}$')
plt.axhline(y=f(0), color='black', linewidth=1.2, linestyle='--')  # Eixo X
plt.axvline(x=0, color='black', linewidth=1.2, linestyle='--')  # Eixo Y
plt.scatter(t, f(t), label='Pontos de amostra')
plt.title('Gráfico Quantidade de Bactérias em (Milhões)', fontsize=16)
plt.xlabel('t')
plt.ylabel('Valores em Milhões')
plt.legend()
plt.show()
```

### Exercício: O Crescimento de uma População de Bactérias

Imagine que uma colônia de bactérias começa com **100 indivíduos** e, em condições ideais, sua população dobra a cada hora.

O número de bactérias $N$ após $t$ horas pode ser modelado por uma função exponencial do tipo $N(t) = C \cdot a^t$.

Baseado nesta informação e na explicação sobre funções exponenciais:

1.  Identifique os valores de $C$ e $a$ para esta situação e escreva a função exponencial $N(t)$ específica que modela o tamanho da colônia após $t$ horas.
2.  Qual será o tamanho da colônia após 3 horas?
3.  Qual o tamanho inicial da colônia ($N(0)$)? Confirme que ele corresponde ao valor informado no início.

```python
def N(t):
    C = 100 # População inicial
    a = 2 # Razão de crescimento
    return C * a ** t 

# Gráfico da função
plt.figure(figsize=(12, 6))

# Definindo o intervalo de x
t = np.linspace(0, 4, 5)
ax = sns.lineplot(x=t, y=N(t), label=r'$N(t) = C \cdot a^t$') 
plt.axhline(y=0, color='black', linewidth=1.2, linestyle='--')  # Eixo X
plt.axvline(x=0, color='black', linewidth=1.2, linestyle='--')  # Eixo Y
plt.scatter(t, N(t), label='Pontos de amostra')
plt.title('Crescimento de uma População de Bactérias', fontsize=16)
plt.xlabel('Horas')
plt.ylabel('N(t)')
plt.legend()

print("A quantidade de bactérias em 3 horas é", N(3))

plt.show()
```

O automóvel é um bem que se desvaloriza muito rapidamente, quando comparado a outros bens. Após a venda, um automóvel novo já sofre uma grande desvalorização. O histórico de um automóvel novo, vendido por $\$ 30.000,00$, apresenta os seguintes valores $(V)$ de mercado, após decorridos os períodos indicados a seguir:


- Ao final de um ano, $R\$ 27.000,00$; 
- Ao final de dois anos, $R\$ 24.300,00$; 
- Ao final de três anos, $R\$ 21.870,00$.

Esses preços seguiram um modelo exponencial que expressa $V$ em função do número n de ano de uso, pela relação $V(n) = V_0 · q^n$, em que $V_0$ é o valor inicial, $q$ é o fator de desvalorização e $n$ é o tempo, em ano, decorrido após a venda.

O valor, em milhar de real, com uma casa decimal, que mais se aproxima do valor de mercado desse carro, ao final de seis anos, é
Alternativas

A)13,7.

B)14,3.

C)14,6.

D)15,9.

E) 17,7. 


```python
def V(n):
    V_0 = 30000 # Valor inicial
    q = 0.1 # Razão de depreciação
    return (V_0 * (1 - q) ** n) # Fórmula do valor após n anos

V(6) # Valor após 6 anos
```
