# O que é uma Função Afim?

Antes de tudo, vale diferenciar dois conceitos importantes:

- **Função Afim** é toda função da forma $ f(x) = ax + b $, onde $ a \neq 0 $ e $ b \in \mathbb{R} $.
- **Função Linear** é um caso particular da função afim onde $ b = 0 $, ou seja, a reta passa pela origem.
- Já a **Função Constante** acontece quando $ a = 0 $, resultando numa reta horizontal.

---

## O que é uma Função Afim?

A função afim é uma das mais importantes da matemática e modela situações onde há uma **relação constante entre duas variáveis**. Sua forma geral é:

$$
f(x) = ax + b
$$

- $a$ é o **coeficiente angular** (indica a inclinação da reta).
- $b$ é o **coeficiente linear** (indica onde a reta corta o eixo $y$).

Ela descreve **retas não verticais**, e aparece em muitos contextos do mundo real, como cálculos de custo, velocidade constante, variações de temperatura, entre outros.

---

## Como ela é escrita?

A forma geral da função afim é:

$$
f(x) = ax + b
$$

- Se $a = 0$, a função vira uma constante (reta horizontal).
- Se $b = 0$, a função vira uma função linear (passa pela origem).
- A função sempre gera uma reta, com **inclinação definida por $a$**.

### Exemplo:

Se $f(x) = 2x + 3$, então:

- $f(0) = 3$
- $f(1) = 5$
- $f(-2) = -1$

---

## Características

- **Gráfico**: Reta inclinada que pode subir ou descer.
- **Crescimento**:
  - **Crescente** se $a > 0$
  - **Decrescente** se $a < 0$
- **Raiz**: O ponto onde a função cruza o eixo $x$, calculado por:

$$
f(x) = 0 \Rightarrow x = -\frac{b}{a}
$$

- **Domínio**: $\mathbb{R}$
- **Imagem**: $\mathbb{R}$

---

## E o gráfico?

O gráfico da função afim é uma **reta**. Sua inclinação depende do valor de $a$, e sua altura inicial depende de $b$.

Por exemplo, se:

$$
f(x) = 2x + 1
$$

A reta **começa em $y = 1$** e **sobe 2 unidades para cada 1 unidade que avança em $x$**.

---

## Aplicações no dia a dia

Funções afins aparecem em diversos contextos que envolvem **crescimento ou decrescimento constante**.

| Situação                                      | Explicação                                                                 | Função afim            |
|----------------------------------------------|---------------------------------------------------------------------------|-------------------------|
| Cálculo de custo fixo + variável             | R$10 fixo + R$2 por item                                                  | $f(x) = 2x + 10$        |
| Salário com valor por hora + adicional fixo  | R$20 por hora + R$100 fixo                                                | $f(x) = 20x + 100$      |
| Temperatura aumentando 2º por hora           | A cada hora a temperatura sobe 2 graus                                    | $f(x) = 2x + T_0$       |
| Distância percorrida em velocidade constante | Um carro andando a 60 km/h                                                | $f(x) = 60x$            |
| Valor de um serviço com taxa inicial         | R$30 fixo + R$5 por km de deslocamento                                    | $f(x) = 5x + 30$        |


## Gráfico comparativo entre os três tipos de funções:

Função Afim: reta inclinada que não passa pela origem (ex: $f(x)=2x+3$).

Função Linear: reta inclinada que passa pela origem (ex: $f(x)=2x$).

Função Constante: reta horizontal (ex: $f(x)=3$).

```python
import matplotlib.pyplot as plt
import numpy as np

# Intervalo de x
x = np.linspace(-10, 10, 400)

# Funções
y_afim = 2 * x + 3        
y_linear = 2 * x         
y_constante = np.full_like(x, 3)  

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(x, y_afim, label="Função Afim: f(x) = 2x + 3", color='steelblue')
plt.plot(x, y_linear, label="Função Linear: f(x) = 2x", color='blueviolet')
plt.plot(x, y_constante, label="Função Constante: f(x) = 3", color='orangered')

# Estética do gráfico
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.title("Gráfico Comparativo: Função Afim, Linear e Constante")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.tight_layout()

plt.show()
```

## Exemplo aplicado: Assinatura mensal com custo inicial

### Enunciado

Uma empresa oferece um serviço de streaming que cobra **R\$50,00 de taxa de adesão** no primeiro mês e **R\$30,00 por mês de assinatura**. 

1. Modelando a função que representa o **custo total** em função da quantidade de meses.
2. Determinando o custo total para 1, 3, 6 e 12 meses.
3. Representando graficamente a função no intervalo de $x = 0$ até $x = 12$.

---

### Solução

#### 1. Modelagem da função

O custo total é composto por uma parte **fixa (adesão)** e uma parte **variável (mensalidade)**. Isso é uma **função afim**:

$$
f(x) = 30x + 50
$$

Onde:

- $x$ = número de meses de assinatura  
- $f(x)$ = custo total em reais  
- $30$ = valor mensal  
- $50$ = taxa fixa de adesão

---

#### 2. Aplicando valores

- Para 1 mês: $f(1) = 30 \cdot 1 + 50 = R\$80,00$  
- Para 3 meses: $f(3) = 30 \cdot 3 + 50 = R\$140,00$  
- Para 6 meses: $f(6) = 30 \cdot 6 + 50 = R\$230,00$  
- Para 12 meses: $f(12) = 30 \cdot 12 + 50 = R\$410,00$

---

#### 3. Gráfico

O gráfico da função é uma reta que:

- Corta o eixo $y$ em $f(0) = 50$  
- Cresce $30$ unidades em $y$ para cada unidade em $x$

A função **não passa pela origem**, pois há uma taxa inicial ($b = 50$).


```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo o intervalo de 0 a 10 meses
x = np.linspace(0, 10, 100)
y = 30 * x + 50

# Pontos importantes
x_pontos = [0, 6, 7, 10]
y_pontos = [30*xi + 50 for xi in x_pontos]

# Criando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=r'$f(x) = 30x + 50$', color='steelblue', linewidth=2)
plt.scatter(x_pontos, y_pontos, color='orangered', zorder=5, linewidth=0.5)

# Anotações
plt.text(6, y_pontos[1] + 5, 'f(6) = R$230', ha='right')
plt.text(7, y_pontos[2] + 5, 'f(7) = R$260', ha='right')

plt.title('Gráfico da Função Afim: f(x) = 30x + 50')
plt.xlabel('Meses (x)')
plt.ylabel('Custo Total (R$)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
