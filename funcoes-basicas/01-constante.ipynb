# O que é uma Função Constante?

A função constante é o tipo mais simples de função matemática. Ela associa qualquer valor real de entrada $x$ a um único valor fixo de saída $c$, ou seja:

$$f(x) = c$$

Independentemente do valor de $x$, a função sempre retorna o mesmo resultado. Seu gráfico é uma reta horizontal paralela ao eixo $x$, localizada na altura $y = c$.

Essa função é fundamental para entender padrões invariáveis, como sistemas que se mantêm estáveis ou grandezas que não variam com o tempo. Apesar de sua simplicidade, possui aplicações em contextos reais como temperatura controlada, salário fixo ou sensores que registram valores constantes.

---

## Como ela é escrita?

A forma geral é:

$ f(x) = c $

- Esse $c$ é um número fixo.
- O $x$ é a entrada (o valor que você escolhe).
- Não importa o valor de $x$, o resultado sempre vai ser o mesmo $c$.

### Exemplo:
$ f(x) = 5 $

- Se você colocar $x = 1$, o resultado é 5.  
- Se colocar $x = 1000$, ainda dá 5.  
- Se colocar $x = -10$, ainda dá 5.

---

## Características

- O gráfico é uma linha reta horizontal.
- O valor de saída não depende de $x$.
- Domínio: $\mathbb{R}$.
- Imagem: $\{c\}$.
- Não possui raiz, exceto se $c = 0$.

---

## E o gráfico?

O gráfico de uma função constante é uma linha reta na horizontal.  
Ela corta o eixo $y$ no ponto $y = c$ e nunca sobe ou desce.

### Exemplos:
- $ f(x) = 5 $ → linha reta na altura 5.  
- $ f(x) = -2 $ → linha reta na altura -2.

---

## Aplicações no dia a dia

| Situação                     | Explicação                                 | Função constante |
|-----------------------------|---------------------------------------------|------------------|
| Preço fixo de estacionamento | R$ 10,00 o dia inteiro                      | $ f(x) = 10 $    |
| Temperatura de um freezer    | Sempre -18°C                                | $ f(x) = -18 $   |
| Trabalho com valor fechado   | R$ 500 por serviço, não importa o tempo     | $ f(x) = 500 $   |

## Exemplo aplicado 1: Controle Ideal da Temperatura (Função Constante)

Imagine que você está trabalhando em um laboratório que realiza um experimento químico sensível à temperatura. Para que a reação ocorra corretamente, a temperatura precisa ser mantida constante em 25°C durante todo o experimento, que tem duração de 8 horas.

Podemos representar essa situação por uma função constante:

$
T(t) = 25
$

onde:
- $T(t)$  é a temperatura em graus Celsius no tempo  $t$ , medido em horas;
- A função indica que a temperatura não varia com o tempo é sempre 25°C.

Essa função representa um controle perfeito e estável de temperatura.

```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalo de tempo (0 a 8 horas)
t = np.linspace(0, 8, 100)
T = np.full_like(t, 25)  # Temperatura constante de 25°C

# Plot gráfico
plt.figure(figsize=(10, 5))
plt.plot(t, T, label="T(t) = 25", color="steelblue", linewidth=2)
plt.title("Temperatura constante no experimento")
plt.xlabel("Tempo (horas)")
plt.ylabel("Temperatura (°C)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.yticks(np.arange(0, 40, 5))
plt.grid(True)
plt.legend()
plt.show()
```

## Exemplo aplicado 2: Falha no Controle da Temperatura

Agora, suponha que ocorra uma falha no sistema de controle térmico do mesmo experimento. Com essa falha, a temperatura começa a oscilar e subir levemente ao longo do tempo, comprometendo a estabilidade do processo químico.

A nova temperatura registrada é modelada pela função:

$$
T_{\text{falha}}(t) = 25 + \sin(t) + \frac{t}{8}
$$

onde:

- $ \sin(t) $ representa oscilações cíclicas no sistema, como variações na resistência elétrica do aquecedor;
- $ \frac{t}{8} $ indica uma tendência de aquecimento gradual com o passar das horas;
- A função mostra que a temperatura não é mais estável, o que pode colocar o experimento em risco.

Esse modelo ajuda a visualizar como **a ausência de constância pode afetar um processo sensível**, e a importância de se manter uma função constante como controle ideal.


```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalo de tempo
t = np.linspace(0, 8, 200)

# Temperatura ideal (constante)
T_constante = np.full_like(t, 25)

# Temperatura com falha simulada
T_falha = 25 + np.sin(t) + (t / 8)

# Plot gráfico
plt.figure(figsize=(10, 6))
plt.plot(t, T_constante, label="Temperatura Ideal: T(t) = 25°C", color="steelblue", linestyle="--", linewidth=2)
plt.plot(t, T_falha, label="Temperatura com Falha", color="orangered")
plt.title("Comparação: Controle Ideal vs Temperatura com Falha")
plt.xlabel("Tempo (horas)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.yticks(np.arange(0, 40, 5))
plt.legend()
plt.show()
```
